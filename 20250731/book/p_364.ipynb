```python
# 선형회귀 : 데이터(x)를 넣으면 직선(y)이 어떻게 변화하는지 그 속에서 규칙을 찾는것
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
# y = ax + b (기울기 + 절편)
# x가 0일 때 y가 0이 아닐 수도 있다는 걸 고려

model = LinearRegression(fit_intercept=True)
model
```

```python
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit = np.linspace(-1,11)
xfit
```

```python
xfit = xfit[:, np.newaxis]
xfit.shape
```

```python
yfit = model.predict(xfit)
yfit
```

```python
plt.scatter(x,y)
plt.plot(xfit,yfit)
```
